```python
import requests
from bs4 import BeautifulSoup

# 섹션 딕셔너리
section_dict = {
    '기후/환경':'climate',
    '사회':'society',
    '경제':'economy',
    '정치':'politics',
    '국제':'world',
    '문화':'culture',
    '생활':'life',
    'IT/과학':'tech',
    '인물':'people'
}

# 요청 헤더 설정
req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
}

# 함수선언
def print_news(section_name):
    # 섹션명으로 URL 생성
    section_key = section_dict.get(section_name)
    if not section_key:
        print(f"'{section_name}'은(는) 유효한 섹션명이 아닙니다.")
        return
    
    url = f'https://news.daum.net/{section_key}'
    print(f"======> {url} {section_name} 뉴스 <======")
    
    # 웹 페이지 요청
    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'
    
    # BeautifulSoup 객체 생성
    soup = BeautifulSoup(res.text, 'html.parser')
    news_list = soup.select('ul.list_newsheadline2 li')
    
    # 각 뉴스 항목에서 링크와 제목 추출
    for li_tag in news_list:
        a_tag = li_tag.find('a')
        if a_tag:
            link = a_tag['href']
            
            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
            if strong_tag:
                title = strong_tag.text.strip()
                print(link)
                print(title)
    print()  # 구분을 위한 빈 줄 출력
# 함수호출
print_news('경제')
print_news('사회')
```
